```python
import chainer
import numpy as np
import chainer.functions as F
import chainer.links as L
from chainer import Variable, serializers
from net import MyChain, MyChain2
import pandas as pd
```

```python
nn = MyChain()
serializers.load_npz('result/nn_epoch_3.npz', nn)
#serializers.load_npz('result/nn_iter_99000_n.npz', nn)
test_X = np.load("test_npy/test_Xxmini.npy").astype(np.float32)
test_t = np.load("test_npy/test_tmini.npy").astype(np.int8)
loss = 0
pred = []
for i in range(0, len(test_t), 100):
    x = test_X[i : i + 100]
    t = test_t[i : i + 100]
    with chainer.no_backprop_mode():
        y = nn(x)
        pred += y.data.copy().tolist()
    l = F.sum(F.sigmoid_cross_entropy(y, t))
    print(i, l)
    loss += l
print(loss * 100 / len(test_t))
```

```python
pred = np.array(pred)
pred.shape
```

```python
F.max(F.sigmoid(pred))
```

```python
df_test = pd.read_csv("GCI_TEAM5/test.csv")
df_test.head()
```

```python
dfc_test = df_test.iloc[:, 1:]
standardized_df_test = (dfc_test - dfc_test.mean()) / dfc_test.std()
final_X = standardized_df_test.as_matrix().astype(np.float32)
final_X.shape
```

```python
pred = []
for i in range(0, len(final_X), 100):
    x = final_X[i : i + 100]
    with chainer.no_backprop_mode():
        y = nn(x)
        pred += y.data.copy().tolist()
pred = F.sigmoid(np.array(pred)).data
pred
```

```python
test_result  = pd.DataFrame()
test_result["data_id"] = df_test["data_id"].as_matrix()
print(df_test["data_id"].as_matrix().shape)
test_result["predict"] = pred
test_result.head()
test_result.to_csv( 'result5.csv', index=False, header=False)
```
